```python
import seaborn as sb
import matplotlib as plt
import pandas as pd
import numpy as np
```

```python
#Seaborn has inbuilt dataset

tips = sb.load_dataset('tips')
tips.head()
```

```python
#Setting default theme

sb.set_theme()
```

```python
#visualize the data
sb.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

```python
#load the iris dataset

iris = sb.load_dataset('iris')
iris.head()
```

```python
#Scatterplot

sb.scatterplot( data=iris,x='sepal_length', y='petal_length', hue='species')
```

```python
titanic = sb.load_dataset('titanic')
titanic.head()
sb.countplot(data=titanic, x='class')
```

```python
#barchart

sb.barplot(x='sex',y='survived',hue='class', data=titanic)
```

```python
#house price

from sklearn.datasets import fetch_california_housing
house_ca = fetch_california_housing()

house = pd.DataFrame(house_ca.data, columns=house_ca.feature_names)
house['price'] = house_ca.target
house.head()
```

```python
#Distribution plot
# sb.displot(house['price'])
sb.distplot(house['price'])
```

```python
#Correlation matrix

#using heatmap
import matplotlib.pyplot as plt

correlation = house.corr()
plt.figure(figsize=(10,10))
sb.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
