```python
# import matplotlib

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
print(x)
print(y)
print(z)
```

```python
#plotting the data
#sine wave

plt.plot(x,y)
plt.show()
```

```python
#build cos
plt.plot(x,z)
plt.show()
```
