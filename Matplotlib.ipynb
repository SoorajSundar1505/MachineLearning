```python
# import matplotlib

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
print(x)
print(y)
print(z)
```

```python
#plotting the data
#sine wave

plt.plot(x,y)
plt.show()
```

```python
#build cos
plt.plot(x,z)
plt.show()
```

```python
#adding title, x and y axis label

plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine')
plt.title('Sine Wave')
plt.show()
```

```python
#Parabola
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
#Parabola in diff symbol
# x=np.linspace(-10,10,100)
# y=x**2
plt.plot(x,y, 'g.')  #green
plt.show()
```

```python
#multiple lines
x=np.linspace(-10,10,100)
plt.plot(x,np.sin(x),'r-')
plt.plot(x,np.cos(x),'g--')
plt.show()
```

```python
#Bar plot
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Hindi', 'Tamil']
people = [100,50,150,30,20]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
#Pie chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

languages = ['English', 'French', 'Spanish', 'Hindi', 'Tamil']
people = [100,50,150,30,20]
ax.pie(people,labels=languages,autopct='%1.1f%%')
# plt.xlabel('LANGUAGES')
# plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
#Scatter plot

x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])

ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')

plt.show()
```
